```python
import datetime
t = datetime.time(9, 30, 45, 100000)
print(t)
```

```python
print(t.hour)
```

```python
dt = datetime.datetime(2021, 10, 28, 12, 30, 45, 100000)
print(dt)
```

```python
# print just the year
print(dt.year)
```

```python
#prints current local datetime with a time zone of none
dt_today = datetime.datetime.today()
# gives option to introduce a timezone
dt_now = datetime.datetime.now()
# utcnow does not signify a timezone-aware datetime
# instead it offers utc info set to now but tcinfo still set to none
dt_utcnow = datetime.datetime.utcnow()
print(dt_today)
print(dt_now)
print(dt_utcnow)
```

```python
dt_gmtime = datetime.datetime.gmtime()
print(dt_gmtime)
```
